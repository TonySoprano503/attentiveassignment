```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Function to plot radar chart
def create_radar_chart(values, metrics, campaign_name):
    # Number of variables we're plotting.
    num_vars = len(metrics)

    # Split the circle into even parts and save the angles
    # so we know where to put each axis.
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

    # The plot is a circle, so we need to "complete the loop"
    # and append the start value to the end.
    values += values[:1]
    angles += angles[:1]

    # Draw the radar chart.
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    ax.fill(angles, values, color=barkbox_blue, alpha=0.25)
    ax.plot(angles, values, color=barkbox_blue, linewidth=2)

    # Labels for each metric.
    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(metrics, fontweight='bold')

    # Title
    ax.set_title(f'Email Campaign Metrics: {campaign_name}', fontsize=16, fontweight='bold', pad=20)

    plt.show()

# Load the Excel file
file_path = 'BarkShop Email Data (Brendan Thomas).xlsx'
campaign_data = pd.read_excel(file_path, sheet_name='Campaign Data')

# Select a campaign
campaign_name_may_treats_sale = 'SALE: May Treats'
filtered_data = campaign_data[campaign_data['Campaign'] == campaign_name_may_treats_sale]

# Extract the necessary metrics
metrics = ['Open Rate', 'Click Rate', 'Conversion Rate']
values = [
    filtered_data['Email Unique Open Rate'].values[0],
    filtered_data['Total Click Rate'].values[0],
    filtered_data['Conversion Rate'].values[0]
]

# BarkBox brand colors
barkbox_blue = "#00AEEF"

# Create the radar chart
create_radar_chart(values, metrics, campaign_name_may_treats_sale)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'BarkShop Email Data (Brendan Thomas).xlsx'
data = pd.ExcelFile(file_path)

# Load data from the 'Journeys Data' sheet
journeys_data = pd.read_excel(file_path, sheet_name='Journeys Data')

# Assuming you have a 'Date' column to add manually for this example
# journeys_data['Date'] = pd.to_datetime(journeys_data['Date'])

# Group by date to get the mean rates
journey_metrics = journeys_data.groupby('Date').agg({
    'Email Total Open Rate': 'mean',
    'Total Click Rate': 'mean',
    'CVR': 'mean'
}).reset_index()

# Plot Open, Click, and Conversion Rates Over Time
plt.figure(figsize=(12, 6))
plt.plot(journey_metrics['Date'], journey_metrics['Email Total Open Rate'], marker='o', label='Open Rate', color='g', linewidth=2)
plt.plot(journey_metrics['Date'], journey_metrics['Total Click Rate'], marker='o', label='Click Rate', color='r', linewidth=2)
plt.plot(journey_metrics['Date'], journey_metrics['CVR'], marker='o', label='Conversion Rate', color='c', linewidth=2)
plt.title('Open, Click, and Conversion Rates Over Time for Journeys', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=14, fontweight='bold')
plt.ylabel('Rate', fontsize=14, fontweight='bold')
plt.legend(fontsize=12)
plt.grid(True)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'BarkShop Email Data (Brendan Thomas).xlsx'
data = pd.ExcelFile(file_path)

# Load data from the 'Campaign Data' sheet
campaign_data = pd.read_excel(file_path, sheet_name='Campaign Data')

# Prepare the data for comparison of Open and Conversion Rates by Campaign
campaign_open_conversion = campaign_data[['Campaign', 'Email Unique Open Rate', 'Conversion Rate']]

# Sorting the data to make the plot more readable
campaign_open_conversion = campaign_open_conversion.sort_values(by='Email Unique Open Rate', ascending=False)

# Plot Comparison of Open and Conversion Rates by Campaign
plt.figure(figsize=(14, 8))
plt.plot(campaign_open_conversion['Campaign'], campaign_open_conversion['Email Unique Open Rate'], marker='o', label='Open Rate', color='b', linewidth=2)
plt.plot(campaign_open_conversion['Campaign'], campaign_open_conversion['Conversion Rate'], marker='o', label='Conversion Rate', color='g', linewidth=2)
plt.title('Comparison of Open and Conversion Rates by Campaign', fontsize=16, fontweight='bold')
plt.xlabel('Campaign', fontsize=14, fontweight='bold')
plt.ylabel('Rate', fontsize=14, fontweight='bold')
plt.legend(fontsize=12)
plt.grid(True)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
